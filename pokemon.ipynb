```python
#importation de bibliothéque:
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
#Créez un nouvel objet data_new:
data_new = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
# Afficher les premières lignes pour vérification
data_new.head()
```

```python
#Normaliser les caractéristiques à l'aide de StandardScaler:
scaler = StandardScaler()
X = scaler.fit_transform(data_new)
print(X)
print(StandardScaler)
```

```python
#Vérifier la forme et le type des données normalisées:
print(X.shape)
print(type(X))
```

```python
#Vérification que la distribution est bien centrée:
print(X.mean(axis=0))
print(X.std(axis=0))
```

```python
#Calcul de la matrice de covariance:
import numpy as np

# Calcul de la matrice de covariance
cov_matrix = np.cov(X, rowvar=False)
print(cov_matrix)
```

```python
# Calcul des valeurs propres et des vecteurs propres:
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("Valeurs propres:", eigenvalues)
print("Vecteurs propres:", eigenvectors)
```

```python
# Créer une liste de paires et trier par ordre décroissant des valeurs propres:
# Créer une liste de paires
eigen_pairs = [(np.abs(eigenvalues[i]), eigenvectors[:, i]) for i in range(len(eigenvalues))]

# Trier par ordre décroissant des valeurs propres
eigen_pairs.sort(key=lambda x: x[0], reverse=True)

print("Valeurs propres triées:", [pair[0] for pair in eigen_pairs])
```

```python
#Variance expliquée et choix du nombre de composantes:
# Calculer le pourcentage de variance expliquée
var_explained = [(i / sum(eigenvalues)) * 100 for i in sorted(eigenvalues, reverse=True)]

# Afficher le pourcentage de variance expliquée
print("Variance expliquée par chaque composante:", var_explained)


```

```python
# Tracer la variance cumulée expliquée:
import matplotlib.pyplot as plt

# Variance expliquée cumulée
var_cumulative = np.cumsum(var_explained)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(var_cumulative)+1), var_cumulative, marker='o', linestyle='--')
plt.xlabel('Nombre de composantes principales')
plt.ylabel('Variance cumulée (%)')
plt.title('Variance cumulée en fonction du nombre de composantes principales')
plt.grid()
plt.show()
```

```python
#Déterminer le nombre de composantes nécessaires pour expliquer au moins 90% de la variance:
# Trouver le nombre minimal de composantes pour expliquer 90% de la variance
nb_components = np.argmax(var_cumulative >= 90) + 1

print(f"Nombre de composantes nécessaires pour expliquer au moins 90% de la variance: {nb_components}")
```

```python
#Script permettant de choisir le nombre de composantes pour un seuil arbitraire:
def choisir_composantes_pour_seuil(seuil):
    return np.argmax(var_cumulative >= seuil) + 1

seuil = 95
composantes = choisir_composantes_pour_seuil(seuil)
print(f"Nombre de composantes pour expliquer {seuil}% de la variance : {composantes}")
```

```python
#Projection des données:
# Créer la matrice de projection avec les deux premières composantes principales:
# Sélectionner les deux premières composantes principales
proj_mat = np.hstack((eigen_pairs[0][1].reshape(-1, 1), eigen_pairs[1][1].reshape(-1, 1)))

# Projeter les données normalisées sur les deux premières composantes
pca_data = X.dot(proj_mat)

# Vérifier les dimensions
print(f"Dimensions de proj_mat : {proj_mat.shape}")
print(f"Dimensions de X : {X.shape}")
print(f"Dimensions de pca_data : {pca_data.shape}")
```

```python
#Visualisation de la projection des données:
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df['Speed'], cmap='viridis', edgecolor='k', s=50)
plt.colorbar(label='Speed')
plt.xlabel('Première composante principale')
plt.ylabel('Deuxième composante principale')
plt.title('Projection des données sur les deux premières composantes principales')
plt.grid()
plt.show()
```
